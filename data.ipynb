```python
def categorize_loan_status(status):
    if status in ['Fully Paid', 'Current', 'In Grace Period', 'Does not meet the credit policy. Status:Fully Paid', 'Issued']:
        return 'Good Debt'
    else:
        return 'Bad Debt'

# Thêm cột mới 'loan_status_category'
df['loan_status_category'] = df['loan_status'].apply(categorize_loan_status)

# Hiển thị DataFrame sau khi thêm cột mới
print(df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df_2007_to_2017 = pd.read_csv("lc_db.csv")

# Count the number of rows for each year
year_counts = df_2007_to_2017['year'].value_counts().sort_index()

# Plot trendline
plt.figure(figsize=(10, 6))
plt.plot(year_counts.index, year_counts.values, marker='o', label='Data')

# Fit linear regression model
X = np.array(year_counts.index).reshape(-1, 1)
y = year_counts.values
model = LinearRegression().fit(X, y)

# Predict for future years
future_years = np.arange(2018, 2025).reshape(-1, 1)
predicted_values = model.predict(future_years)

# Plot regression line for future years
plt.plot(future_years, predicted_values, color='green', linestyle='--', label='Predicted Values')

# Calculate regression line for historical data
regression_line = np.polyfit(year_counts.index, year_counts.values, 1)
plt.plot(year_counts.index, np.polyval(regression_line, year_counts.index), color='red', label='Historical Regression Line')

# Add value labels to each data point
for x, y in zip(year_counts.index, year_counts.values):
    plt.annotate(str(y), xy=(x, y), xytext=(10, -10), textcoords='offset points')

# Add value labels to each predicted data point
for x, y in zip(future_years.flatten(), predicted_values):
    plt.annotate(f'{y:.2f}', xy=(x, y), xytext=(10, 5), textcoords='offset points', color='green')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Trendline from 2007 to 2024 with Predictions')
plt.xticks(np.arange(year_counts.index.min(), 2024, step=2, dtype=int))
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Tạo DataFrame mẫu
data_etl = pd.read_csv("lc_db.csv")

# Lọc dữ liệu từ 2007 đến 2017
df_2007_to_2017 = data_etl[(data_etl['year'] >= 2007) & (data_etl['year'] <= 2017)]

# Đếm số lượng hàng cho mỗi năm
year_counts = df_2007_to_2017['year'].value_counts().sort_index()

# Chuẩn bị dữ liệu cho mô hình hồi quy tuyến tính
X = year_counts.index.values.reshape(-1, 1)
y = year_counts.values

# Khởi tạo và huấn luyện mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X, y)

# Mở rộng khoảng thời gian cho đến năm 2025
future_years_extended = np.arange(2018, 2026).reshape(-1, 1)

# Dự đoán giá trị cho các năm
predicted_values_extended = model.predict(future_years_extended)

# Hiển thị dữ liệu và dự đoán trên biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(year_counts.index, year_counts.values, marker='o', label='Data')
plt.plot(future_years_extended.flatten(), predicted_values_extended, linestyle='--', marker='o', label='Predictions')

# Đặt nhãn cho các điểm dữ liệu lịch sử
for x, y in zip(year_counts.index, year_counts.values):
    plt.annotate(str(y), xy=(x, y), xytext=(10, -10), textcoords='offset points')

# Đặt nhãn cho các điểm dự đoán
for x, y in zip(future_years_extended.flatten(), predicted_values_extended):
    plt.annotate(f'{y:.2f}', xy=(x, y), xytext=(10, 5), textcoords='offset points', color='green')

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Trendline from 2007 to 2025 with Predictions')

# Đặt nhãn trục x với các giá trị nguyên của năm
plt.xticks(np.arange(year_counts.index.min(), 2026, step=2, dtype=int))

plt.grid(True)
plt.legend()
plt.show()
```
